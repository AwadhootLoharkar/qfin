```python
import random
from datetime import datetime, timedelta

def generate_series():
    # Constants
    no_of_months = [1, 3, 6, 9, 12, 18, 24, 36]
    stocks_list = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'NVDA', 'META', 'TSLA', 'AVGO', 'ASML', 'COST',
                   'ADBE', 'PEP', 'AMD', 'CMCSA', 'NFLX', 'INTU', 'TXN', 'QCOM', 'PYPL', 'AMGN',
                   'SBUX', 'CHTR', 'MRNA', 'AMAT', 'MU', 'ADP', 'ADI', 'BKNG', 'LRCX', 'VRTX',
                   'KLAC', 'REGN', 'MRVL', 'ILMN', 'GILD', 'ISRG', 'DXCM', 'ROST', 'ORLY']
    start_date_range = datetime(2018, 1, 1)
    end_date_range = datetime(2020, 1, 1)

    # Helper function to add months to a date
    def add_months(start_date, months):
        end_date = start_date + timedelta(days=months*30)
        return end_date


    no_of_list = 3
    no_of_stocks = 10
    series_list = []
    for _ in range(no_of_list):
        selected_stocks = random.sample(stocks_list, no_of_stocks)
        start_date = start_date_range + timedelta(days=random.randint(0, (end_date_range - start_date_range).days))
        duration_months = random.choice(no_of_months)
        end_date = add_months(start_date, duration_months)
        
        if end_date > end_date_range:
            end_date = end_date_range
        
        series = {
            'stocks': selected_stocks,
            'start_date': (start_date.year, start_date.month, start_date.day),
            'end_date': (end_date.year, end_date.month, end_date.day),
            'portfolio_total_value': 1000000,
            'budget': 5,
            'risk_factor': 0.5,
            'confidence_level': 0.95,
            'method': '',
            'solver': ''
        }
        series_list.append(series)

    return series_list

# Generate the series
series_output = generate_series()

# Output the series in the desired format
print("jobs = [")
for idx, series in enumerate(series_output):
    print("    { # series", idx+1)
    for key, value in series.items():
        if isinstance(value, list):
            print(f"        '{key}': {value},")
        else:
            print(f"        '{key}': {value},")
    print("    },")
print("]")
```
